```
df = pd.read_csv('input.csv')
cols = list(df.columns.values)[1:]
df.loc[:, cols] = df.loc[:, cols].cumsum()

strategy_list = []
x = []
lists = list(df)[1:]
dates = list(df['Date'])
for dt in dates:
    dt = str(dt)[:4] + '-' + str(dt)[4:]
    dt = dt[:7] + '-' + dt[7:]
    x.append(dt)

for i, v in enumerate(lists):
    strategy_list.append(
        go.Scatter(
            x=x,
            y=list(df[v]),
            name=v,
            mode="lines+markers"
        )
    )

layout = go.Layout(
    title='Cummulative Chart',
    xaxis=dict(
        showticklabels=True,
    ),
    yaxis=dict(
        showticklabels=True,
    )
)
fig = go.Figure(data=strategy_list, layout=layout)
py.iplot(fig, filename='cummulative-lines')
```
